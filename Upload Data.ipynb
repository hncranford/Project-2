```python
class census(Base):
    __tablename__="citystate"
    ID = Column(Integer, primary_key = True)
    city = Column(String)
    city_ascii = Column(String)
    state_id = Column(String)
    state_name = Column(String)
    county_fips = Column(String)
    county_name = Column(String)
    lat = Column(Float)
    lng = Column(Float)
    population = Column(Float)
    density = Column(Float)
    source = Column(String)
    military = Column(String)
    incorporated= Column(String)
    timezone = Column(String)
    ranking = Column(Float)
    zips = Column(Float)
    
    
```

```python
class census(Base):
    __tablename__="census"
    Rank = Column(Integer, primary_key = True)
    Geographic_Area = Column(String)
    Census = Column(Float)
    Estimates_Base = Column(Float)
    Pop2010 = Column(Float)
    Pop2011 = Column(Float)
    Pop2012 = Column(Float)
    Pop2013 = Column(Float)
    Pop2014 = Column(Float)
    Pop2015 = Column(Float)
    Pop2016 = Column(Float)
    Pop2017 = Column(Float)
    Pop2018 = Column(Float)
    Pop2019 = Column(Float)
```

```python
class analyst(Base):
    __tablename__="analyst"
    Ranks = Column(Integer, primary_key = True)
    Job_Title = Column(String)
    Salary_Estimate = Column(String)
    Job_Description = Column(String)
    Rating = Column(Float)
    Company_Name = Column(String)
    Location = Column(String)
    Headquarters = Column(String)
    Size = Column(String)
    Founded = Column(Float)
    Type_of_ownership = Column(String)
    Industry = Column(String)
    Sector = Column(String)
    Revenue = Column(String)
    Competitors = Column(String)
    Easy_Apply = Column(String)
```

```python
database_path = "postgresql://postgres:postgres@localhost:5432/DataAnalysis"
engine = create_engine(database_path)
conn = engine.connect()
```

```python
Base.metadata.create_all(engine)
```
